## Variables

### Rules for naming Python variables:
- Must start with a letter or an underscore character.
- Cannot start with a number.
- Can only contain alpha-numeric characters and underscores (A-Z, a-z, 0-9, and _ ).
- Case-sensitive (`abc`, `Abc` and `ABC` different variables).

````python
name = 'Diptangsu Goswami'
age = 21
height = 182.88  # cm
````

- You can't declare variables in Python.
- A variable gets created when a value is assigned to it.
- Python is a <font color="blue">dynamically typed</font> language.

````python
print(name)
print(age)
print(height)
print()
print(f'{name} is {age} years old and is {height} centimeters tall')
````

### Swapping two values (in python)

````python
m, n = 20, 40
print('m =', m)
print('n =', n)

print('\nSwapping m and n')
m, n = n, m

print('m =', m)
print('n =', n)
````

## A brief overview about what statically and dynamically typed mean

### Statically typed
- The types of variables are checked at compile time
- The programmer must specify the type of every variable


- A considerable amount of checks can be performed by the compiler
- Reduces trivial bugs

### Dynamically typed
- The types of variables are deduced during run time
- The programmer doesn't have to specify the types of variables


- Enables a programmer to write code quicker

#### The types of variables can be changed

````python
print(name)
print(age)
````

````python
name = 123
age = 'Twenty one years'
````

````python
print(name)
print(age)
````

## Python3 print() function extras

A `print()` function call will print whatever is passed to it and move the cursor to the next line.

````python
print(1000)
print(2000)
print('foo')
````

We can pass an optional keyword argument `end` to the `print` function to override this property. This allows us to print something and keep the cursor in the same line.

````python
print(1000, end='')
print(2000, end='')
print('foo')
print('bar')
````

````python
print(1000, end='#')
print(2000, end='@')
print('foo', end='bar')
````
