```python
import pandas as pd
df = pd.read_csv('Occupation.txt', sep=',', header=None)
df.columns = ['Name','Occupation']
```

```python
xx = df['Occupation'] == "Doctor"
df_doctor = df[df['Occupation'] == "Doctor"]
df_doctor
```

```python
df_actor = df[df['Occupation'] == "Actor"]
df_actor
```

```python
df_professior = df[df['Occupation'] == "Professor"]
df_professior
```

```python
df_singer = df[df['Occupation'] == "Singer"]
df_singer
```

```python
# ná»‘i 2 df = concat
frames = [df_doctor, df_professior]
result = pd.concat(frames, axis=1)
#inplace = true
x = result.drop(columns=["Occupation"])
x
```

```python
frame1 = [df_doctor, df_professior, df_singer, df_actor]
res = pd.concat(frame1, axis=1)
res.drop(columns=['Occupation'])
```

```python
df_doctor.reset_index(drop=True, inplace=True)
df_professior.reset_index(drop=True, inplace=True)
df_singer.reset_index(drop=True, inplace=True)
df_actor.reset_index(drop=True, inplace=True)
df_all = pd.concat([df_doctor,df_professior,df_singer,df_actor], axis =1)
df_all.drop(columns=['Occupation'], inplace=True)
#fillna change NaN
df_all.fillna(value= "-", inplace=True)
df_all
```
